# Exploratory Data Analysis

Qus1. What is the overall sales trend?

```python
df['order_date'].min()
```

```python
df['order_date'].max()
```

```python
#Getting month year from dataset
df['month_year']=df['order_date'].apply(lambda x: x.strftime('%Y-%m'))
df['month_year']
```

```python
df.groupby('month_year').sum()['sales']
```

```python
#Grouping month year trend
df_trend=df.groupby('month_year').sum()['sales'].reset_index()
```

```python
#Setting the figure size
plt.figure(figsize=(16,6))
plt.plot(df_trend['month_year'], df_trend['sales'], color='#b80045')
plt.xticks(rotation='vertical',size=8)
plt.show()
```

Qus2. Which are top 10 products by sales?

```python
df.head()
```

```python
#Grouping by product sales
prod_sales=pd.DataFrame(df.groupby('product_name').sum()['sales'])
```

```python
most_selling_prod=prod_sales.sort_values('sales',ascending=False)
```

```python
#Top 10 product by sales
most_selling_prod.head(10)
```

Qus3. Which are the most selling products?

```python
#Grouping product name
df.groupby('product_name').sum()
```

```python
#most selling products i.e Quantity
heighest_selling_prod=pd.DataFrame(df.groupby('product_name').sum()['quantity'])
```

```python
#Sorting by heighest
best_selling_prod=heighest_selling_prod.sort_values('quantity',ascending=False)
```

```python
best_selling_prod[:10]
```
